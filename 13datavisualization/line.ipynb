```python
import matplotlib.pyplot as plt
import numpy as np
# pyplot is object
```

```python
x=np.arange(10)
y1=x**2
y2=2*x+3
print(x)
print(y1)
print(y2)
```

```python
themes=plt.style.available
themes
```

```python
# line plot
plt.style.use('seaborn')
plt.plot(x,y1,color='red',label='y1',marker='o')
plt.plot(x,y2,label='y2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('prices')
plt.show()
plt.plot(x,y2)
plt.show()
```
